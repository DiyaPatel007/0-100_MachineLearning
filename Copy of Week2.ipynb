Linear Regression

```python
#Write Code for Logistic Regression here
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

X = cancer_dataset.iloc[:,1:-1].values  #Features
y = cancer_dataset.iloc[:,-1].values    #Target

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2)

model = LogisticRegression(max_iter=10000) # Increased max_iter in case the model needs more iterations to converge
model.fit(X_train,y_train)

print(f'LR : {model.score(X_test,y_test)}')
```

SVM(Support Vector Machines)

```python
#Write Code for SVM here
import pandas as pd
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split

X = cancer_dataset.iloc[:,1:-1].values #Features
y = cancer_dataset.iloc[:,-1].values    #Target

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2)

clf = SVC(kernel= 'linear',C=7)
clf.fit(X_train,y_train)

print(f'SVM : {clf.score(X_test,y_test)}')
```

# **LINEAR REGRESSION**

```python
#For more information for the dataser refer: https://archive.ics.uci.edu/dataset/477/real+estate+valuation+data+set
real_estate_dataset = pd.read_excel('/content/Real estate valuation data set.xlsx')
```

```python
real_estate_dataset.head()
```

```python
real_estate_dataset.tail()
```

```python
# Code for Linear Regression

import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

df = pd.DataFrame(real_estate_dataset)

X = df.loc[:, 'X1 transaction date':'X6 longitude']
y = df.loc[:, 'Y house price of unit area']


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=20)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred_train = model.predict(X_train)

mse_train = mean_squared_error(y_train, y_pred_train)
r2_train = r2_score(y_train, y_pred_train)

print(f"Mean Squared Error: {mse_train}")
print(f"R^2 Score: {r2_train}")

plt.scatter(y_train, y_pred_train)
plt.xlabel('Actual Data')
plt.ylabel('Predicted Data')
plt.title('Linear Regression for Trained Data')
plt.show()

y_pred_test = model.predict(X_test)

mse_test = mean_squared_error(y_test, y_pred_test)
r2_test = r2_score(y_test, y_pred_test)

print(f"Mean Squared Error: {mse_test}")
print(f"R^2 Score: {r2_test}")

feature = 'X2 house age'
plt.scatter(y_test, y_pred_test)
plt.xlabel('Actual Data')
plt.ylabel('Predicted Data')
plt.title('Linear Regression for Tested Data')
plt.show()
```
